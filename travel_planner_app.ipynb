```python
import streamlit as st
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage

# ‚úÖ Securely Fetch API Key from Streamlit Secrets
GOOGLE_API_KEY = 'AIzaSyA3fhOqw-XuOQ3DwgSGzJqttcr-QOVHRbs'

# ‚úÖ Function to fetch AI-generated travel options
def get_travel_options(source, destination):
    system_prompt = SystemMessage(
        content="You are an AI-powered travel assistant. Provide multiple travel options (cab, train, bus, flight) with estimated costs, duration, and relevant travel tips."
    )
    user_prompt = HumanMessage(
        content=f"I am traveling from {source} to {destination}. Suggest travel options with estimated cost, duration, and important details."
    )

    # ‚úÖ Initialize AI model
    llm = ChatGoogleGenerativeAI(model="gemini-1.5-pro", google_api_key=GOOGLE_API_KEY)

    try:
        response = llm.invoke([system_prompt, user_prompt])
        return response.content if response else "‚ö†Ô∏è No response from AI."
    except Exception as e:
        return f"‚ùå Error fetching travel options: {str(e)}"

# ‚úÖ Streamlit UI
st.title("üöÄ AI-Powered Travel Planner")
st.markdown("Enter your travel details to get AI-generated travel options including cost estimates and travel tips.")

# ‚úÖ User Inputs
source = st.text_input("üõ´ Enter Source Location", placeholder="E.g., New York")
destination = st.text_input("üõ¨ Enter Destination", placeholder="E.g., Los Angeles")

if st.button("üîç Find Travel Options"):
    if source.strip() and destination.strip():
        with st.spinner("üîÑ Fetching best travel options..."):
            travel_info = get_travel_options(source, destination)
            st.success("‚úÖ Travel Recommendations:")
            st.markdown(travel_info)
    else:
        st.warning("‚ö†Ô∏è Please enter both source and destination locations.")
```
